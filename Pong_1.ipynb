```python
import turtle   # Importing library to be used.


# Objects for game.
win = turtle.Screen() 
win.title("Game")
win.bgcolor("black")
win.setup(height = 600, width = 800)
win.tracer(0)


paddle_1 = turtle.Turtle() 
paddle_1.speed(0)
paddle_1.shape("square")
paddle_1.color("yellow")
paddle_1.penup()  
paddle_1.shapesize(stretch_wid = 5, stretch_len = 1)
paddle_1.goto(-375,0)

paddle_2 = turtle.Turtle() 
paddle_2.speed(0)
paddle_2.shape("square")
paddle_2.color("red")
paddle_2.penup()
paddle_2.shapesize(stretch_wid = 5, stretch_len = 1)
paddle_2.goto(375,0)

score_1 = 0
score_2 = 0

ball = turtle.Turtle()
ball.shape("circle")
ball.speed(1)
ball.color("white")
ball.penup()
ball.goto(0,0)
ball.cx = 2
ball.cy = 2  

pen = turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0,260)
pen.write("Player A: 0  Player B: 0", align="center", font=("Comic Sans MS", 20, "normal"))


# Functions for movements.
def paddle_1_up():
    y = paddle_1.ycor()
    y+=20
    paddle_1.sety(y)  
    
def paddle_1_down():
    y = paddle_1.ycor()  
    y-=20
    paddle_1.sety(y) 
    
def paddle_2_up():
    y = paddle_2.ycor()  
    y+=20
    paddle_2.sety(y) 
    
def paddle_2_down():
    y = paddle_2.ycor()  
    y-=20
    paddle_2.sety(y)  
    
    
# Taking direction input from user.
win.listen()   
win.onkeypress(paddle_1_up,"w")   
win.onkeypress(paddle_1_down,"s") 
win.onkeypress(paddle_2_up,"Up") 
win.onkeypress(paddle_2_down,"Down")  
    

while True:
    win.update()
    
    # Moving the ball.
    ball.setx(ball.xcor() + ball.cx)
    ball.sety(ball.ycor() + ball.cy)
    
    
    if ball.ycor()>290:
        ball.sety(290)
        ball.cy*=-1
        
    if ball.ycor()<-290:
        ball.sety(-290)
        ball.cy*=-1
        
    if ball.xcor()>390:
        ball.goto(0, 0)
        ball.cx*=-1
        score_1+=1
        pen.clear() 
        pen.write("Player A: {}  Player B: {}".format(score_1, score_2), align="center", font=("Comic Sans MS", 20, "normal"))
        
    if ball.xcor()<-390:
        ball.goto(0, 0)
        ball.cx*=-1
        score_2+=1
        pen.clear()
        pen.write("Player A: {}  Player B: {}".format(score_1, score_2), align="center", font=("Comic Sans MS", 20, "normal"))
        
    if (ball.xcor()>365 and ball.xcor()<375) and (ball.ycor()<paddle_2.ycor()+40 and ball.ycor()>paddle_2.ycor()-70) :
        ball.setx(340)
        ball.cx*=-1

        
    if (ball.xcor()<-365 and ball.xcor()>-375) and (ball.ycor()<paddle_1.ycor()+70 and ball.ycor()>paddle_1.ycor()-50) :
        ball.setx(-340)
        ball.cx*=-1
        
```
